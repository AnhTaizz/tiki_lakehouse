```python
from pyspark.sql import SparkSession

# Khởi tạo cỗ máy Spark
spark = SparkSession.builder \
    .appName("Day 13 - SQL & Time Travel") \
    .getOrCreate()

print("Đã khởi động Spark thành công!")
```

```python
# Đọc dữ liệu từ file CSV có sẵn trong thư mục notebooks của bạn
df = spark.read.csv("/home/jovyan/work/notebooks/tiki_skincare_ready.csv", header=True, inferSchema=True)

# Ghi dữ liệu vào bảng Iceberg (đây là lúc Spark ghi lại 'địa chỉ' vào Postgres mới)
df.writeTo("local_catalog.tiki.products").createOrReplace()

print("Đã đăng ký lại bảng Tiki thành công!")
```

```python
spark.sql("SELECT * FROM local_catalog.tiki.products LIMIT 5").show()
```
