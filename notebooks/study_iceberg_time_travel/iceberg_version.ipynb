```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Day 13 - Time Travel").getOrCreate()

spark.sql("SELECT * FROM local_catalog.tiki.products.history").show(truncate=False)
```

```python
print("1. Lịch sử các lần thay đổi dữ liệu:")
spark.sql("SELECT * FROM local_catalog.tiki.products.history").show(truncate=False)
```

```python
print("2. Tiến hành xóa dữ liệu...")
spark.sql("DELETE FROM local_catalog.tiki.products WHERE brand_name = 'La Roche-Posay'")
print("Oops! Đã lỡ tay xóa sạch hàng của La Roche-Posay!")

print("Kiểm tra lại bảng xếp hạng Top 5:")
spark.sql("""
    SELECT brand_name, count(*) as so_luong_sp
    FROM local_catalog.tiki.products 
    GROUP BY brand_name 
    ORDER BY so_luong_sp DESC
    LIMIT 5
""").show(truncate=False)
```
